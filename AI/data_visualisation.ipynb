**Ways to handle imbalance data**
- UpSampling
    - Increase number of minority classes instances e.g. SMOTE (Synthetic Minority Over-sampling Technique)
    - GANs  for images, but since we already have numerical matrix, we do not have to use GAN, as that is computionally expensive

- DownSampling
    - Decrease number of majority classes instances


**Which one to apply and why?**
- DownSampling Benefits:
    - Computionally cheaper
    - Not using any synthetic data, that is better. because if UpSampling creates non-realistic synthetic data , then it would probably decrease the accuracy of the machine learning model
    - better for large datasets

- DownSampling Disadvantages:
    - Information loss



- UpSampling Benefits:
    - SMOTE-KNN used by Leveraging MediaPipe and YOLO Keypoint Detection in Ensemble Approaches for Workout Pose Recognition [Academic article]
    - According to Leveraging MediaPie and... using SMOTE can be considered as data augmentation to increase diversity and generalisation in dataset

```python
def Preprocess_data(dataframe,columns_to_flatten):
    final_df=dataframe.copy()
    # Expanding each column into 3 separate columns (x, y, z) and appending it to the final dataframe.
    for column in columns_to_flatten:
        print(np.vstack(dataframe[column].str.strip('[]').str.split()).astype(float))
        expanded_df=pd.DataFrame(np.vstack(dataframe[column].str.strip('[]').str.split()).astype(float), 
                           columns=[column+'_x', column+'_y', column+'_z'],
                           index=dataframe.index)
        new_df = pd.concat([dataframe.drop(column, axis=1), expanded_df], axis=1)
        for new_column in new_df.columns:
            final_df[new_column] = new_df[new_column]
        # final_df.drop(column, axis=1,inplace=True)

    return final_df.drop(columns=columns_to_flatten,axis=1)

def Return_X_y(dataframe,columns_to_delete):
    X=dataframe.drop(columns=columns_to_delete)
    y=dataframe['label']
    return X,y

final_df=Preprocess_data(training_dataset,['left_shoulder',
       'right_shoulder', 'left_elbow', 'right_elbow', 'left_wrist',
       'right_wrist', 'left_pinky', 'right_pinky', 'left_index', 'right_index',
       'left_thumb', 'right_thumb', 'left_hip', 'right_hip', 'left_knee',
       'right_knee', 'left_ankle', 'right_ankle', 'left_heel', 'right_heel',
       'left_foot_index', 'right_foot_index'])

X_train, y_train = Return_X_y(final_df,['label','muscle group','WorkoutLabel','image'])
```

**Feature Engineering**
- Remove features with high percentage missing value
- Drop features that has same value most of the time [Low Variation features]
    - Drop for sure the features with zero variation (unary)
- Remove features that are highly correlated with each other, remove one of these features
    - Remove the one that is least correlated with each y_target
- Drop features that have very low correlation with y_target
